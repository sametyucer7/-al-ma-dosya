```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data = { "year" : np.arange(2010,2021) ,
        "satış" : np.random.randint(100,500,11),
        "masraf" : np.random.randint(50,400,11)}

df = pd.DataFrame(data)

regions = ["kuzey","güney","doğu","batı"]

df["Bölge"] = np.random.choice(["kuzey","güney","doğu","batı"],df.shape[0]) # SHAPE İLE KAÇ SÜTÜN VE SATIR OLDUĞUNU ÖĞRENİYORUZ [0] İLEDE 11,3 OLAN SHAPEDEN 0 İNDEXİ İLE 11 E ULAŞIYORUZ

df
```

```
plt.plot(df["year"],df["satış"],
         color = "red",
         marker="o",
         linestyle="--",
         label="satış",
         markerfacecolor="black")

plt.plot(df["year"] , df["masraf"],label="masraf",
         color="green",marker="x",linestyle="--")

plt.title("yıllık gelir gider grafiği") # BAŞLIK ATMAK İÇİN

plt.xlabel("year", fontsize = 20)
plt.ylabel("tl",fontsize = 20)

plt.legend()
plt.show()
```

```
sales_by_region = df.groupby("Bölge")["satış"].sum()


plt.bar(sales_by_region.index,sales_by_region.values)

plt.xlabel("Satış")
plt.ylabel("BÖLGE")
plt.title("BÖLGELERİN TOPLAM  SATIŞI")
plt.show()
```

```
# pandas cut : bin arası mesafe eşit

# sales_bins = pd.cut(df["satış"],bins=10)

#pandas qcut : bin arası mesafe eşit değil

# bin : numerik datanın grubu olarak düşünülebilir

# (113.999 , 175.0) --> bin i ifade eder

sales_bins = pd.qcut(df["satış"],q=10)

bin_counts = sales_bins.value_counts().sort_index()

for aralık in  bin_counts.index:
  merkez = (aralık.left + aralık.right) /2
  print(merkez)
```

---PLT.HİST()

bins : grup sayısını belirler

range : binlerin alt ve üst limitlerini belirler

density : TRUE olursa grafik yoğunluk belirtir  False olursa frekans belirtir

```
plt.hist( df["satış"] , bins = 10 , color = "blue" , width = 5 , label="Satış frekansı")
plt.legend()
plt.show()
```
