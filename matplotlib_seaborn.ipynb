```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
plt.rcParams["figure.figsize"] = (5,5) # ÇERÇEVENİN BOYUTUNU KALICI YAPMA

plt.rcParams["xtick.labelsize"] = 12 # X GRAFİĞİNDEKİ DEĞERLERİN BOYUTU

plt.rcParams["ytick.labelsize"] = 12 # Y GRAFİĞİNDEKİ DEĞERLERİN BOYUTU

plt.rcParams["axes.labelsize"] = 20  # X VE Y GRAFİĞİ BAŞLIKLARI BOYUTU
```

```
data_example = sns.load_dataset("penguins")
data_example.head()
```

```
data_male=data_example[data_example["sex"] == "Male"] ["flipper_length_mm"]

data_female=data_example[data_example["sex"] == "Female"] ["flipper_length_mm"]

sns.histplot(data = data_female , label= "female", color="pink",alpha=0.5)
sns.histplot(data = data_male , label="male",color="blue" ,alpha=0.5)
plt.legend()
plt.show()


###### BUNUN YER,NE SEABPRNDA TEKK SATIRDA YAPILABİLCEK BİR KOMUT VAR
```

```
########## TEK SATIRDA HALİ

sns.histplot(data = data_example , x ="flipper_length_mm" , hue="sex", alpha=0.5 )

# LABEL İ KENDİ KOYDU
```

---- **SEABORN İLE SCATTERPLOT**


```
sns.scatterplot(data=data_example, x = "bill_length_mm" , y = "bill_depth_mm" , hue = "species")
```

---- **SEABORN İLE BARPLOT**

```
grup_data = data_example.groupby(["species","sex"])["body_mass_g"].mean().reset_index()
```

```
sns.barplot(data = grup_data ,
            x = "species" ,
            y = "body_mass_g",
            hue="sex")
```

---- **SUBPLOT OLUŞTURMA**

```
fig , axes = plt.subplots( 2 , 2 , figsize = (7,7) )
sns.scatterplot(data=data_example
                , x = "bill_length_mm"
                , y = "bill_depth_mm"
                , hue = "species"
                , ax = axes[1,1])

# (0,0) - (0,1 ) - (1,0 ) - (1,1)
 # sol üst - sağ üst - sol alt - sağ alt

 ##### dir(axes()) YOLU İLE AXESİN ALDIĞI FONKSİYONLARI GÖSTERİR

axes[0,0].set_title("Bill length_mm vs bill_depth_mm")
```

```
sns.histplot(data = data_example,
             x = "bill_length_mm",
             ax=axes[1,0],
             hue="species"
             )
```
