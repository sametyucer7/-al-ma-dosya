----**MATPLOTLİB**

```
import matplotlib.pyplot as plt
import numpy as np
```

```
#önce çerçeve oluşturulur sonra veri yerleştirilir

plt.figure(figsize=(5,5))     # TABLONUN BOYUTUNU AYARLAMAYA YARAR

plt.show() # ÇERÇEVE YAPILDIYSA ONU GÖSTERİR
```

**---- plt.plot fonksiyonu**


```
x = [0,1,2,3,4,5,6]
y = [0,2,4,6,8,10,12]
plt.plot(x,y,"r--")
plt.show()
```

```
x = np.arange(6)
y= 4*x - 2

plt.figure(figsize=(5,5))

plt.plot(x,y,"g--")
plt.show()
```

---- **LİNİER OLMAYAN FONKSİYONLAR**


```
x = np.arange(100) / 10
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```
x = np.arange(6)
y1 = x * 2
y2 = x ** 2
plt.figure(figsize=(5,5))
# TEK ÇERÇEVEDE 2 GRAFİK YAZMA YOLU
plt.plot(x,y1,"r",label="(f) = 2x")
plt.plot(x,y2,"g",label="(f) = x ** 2  ")

plt.legend(loc="lower right") # BU plt.legend KOMUTU ÇERÇEVEDE GİRİLEN GRAFİĞİ AYIRT ETME PENCERESİNİ AKTİF EDER .

# plt.legend(loc="best") TAHMİNİ EN İYİ KONUMA YERLEŞTİRİLİR.

plt.show()
```

---- **plt.scatter**

```
x = [0,1,2,3,4,5,6]

y = [0,2,4,6,8,10,12]

plt.figure(figsize=(5,5))

plt.scatter(x,y,color="blue") #### plt.plot()  yerine plt.scatter() yazınca grafik yerine direkt x ler ile y lerin kesiştiği yerlere nokta koyar. ama daha okunaklı grafik için altına plotu da yazmak mantıklı olur.

plt.plot(x,y,"red")

plt.show()
```

```
x = np.arange(100) / 10
y = np.cos(x)
z = np.sin(x)

plt.scatter(x , y,color="red",label="cosx")
plt.plot(x , z,color="green",label="sinx")

plt.legend()

plt.show()
```

```
########### SCATTER OLMADAN DEĞERLERE NOKTA KOYMA

x = [0,1,2,3,4,5,6]

y = [0,2,4,6,8,10,12]

plt.plot(x , y , color="green" , label="sinx" , linestyle = "--", marker="o" , markersize = 10 , alpha = 1) # ALPHA = OPAKLIK SAYDAMLIK İÇİN KULLANILIR

plt.legend()

plt.show()
```
